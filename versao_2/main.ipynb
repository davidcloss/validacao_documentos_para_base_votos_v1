```python
from validate_docbr import CPF, CNPJ
import pandas as pd
```

```python


def validar_documento(documento):
    cpf = CPF()
    cnpj = CNPJ()

    documento = str(documento).strip()  # Converte para string e remove espaços
    documento = documento.replace(".", "").replace("-", "").replace("/", "")

    # Adiciona zeros à esquerda para ter o comprimento correto
    if len(documento) <= 11:
        documento = documento.zfill(11)
        if len(documento) == 11 and cpf.validate(documento):
            return 'FISICA'
    elif 12 <= len(documento) <= 14:
        documento = documento.zfill(14)
        if len(documento) == 14 and cnpj.validate(documento):
            return 'JURIDICA'

    # Tentativa de validar como CPF
    if len(documento) == 11 and cpf.validate(documento):
        return 'FISICA'
    
    # Tentativa de validar como CNPJ
    if len(documento) == 14 and cnpj.validate(documento):
        return 'JURIDICA'

    return "Documento inválido"

```
